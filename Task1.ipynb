```python
import pandas as pd
```

```python
#loading a dataset
df = pd.read_csv(r"C:/Users/ADMIN/Downloads/archive (3)/sales_data_sample.csv", encoding='latin1')
```

```python
df.head()
```

```python
df.isnull().sum() #checking missing values
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['ADDRESSLINE2'])
```

```python
df['STATE'] = df['STATE'].fillna("Unknown")
```

```python
df['TERRITORY'] = df['TERRITORY'].fillna("Unknown")
```

```python
df['POSTALCODE'] = df['POSTALCODE'].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_") #standarized column names
```

```python
df.head()
```

```python
df['orderdate'] = pd.to_datetime(df['orderdate'], errors='coerce') #converting ORDERDATE to proper date format
```

```python
#numeric_cols = ['quantityordered', 'priceeach', 'orderlinenumber', 'sales']

#for col in numeric_cols:
#    df[col] = pd.to_numeric(df[col], errors='coerce')
df['quantityordered'] = pd.to_numeric(df['quantityordered'])
df['priceeach'] = pd.to_numeric(df['priceeach'])
df['orderlinenumber'] = pd.to_numeric(df['orderlinenumber'])
df['sales'] = pd.to_numeric(df['sales'])
```

```python
df['sales'].isnull() 

df['quantityordered'].notnull() 
 
df['priceeach'].notnull()
```

```python
mask = df['sales'].isnull() & df['quantityordered'].notnull() & df['priceeach'].notnull()
mask.sum()  #sales has no missing values
```

```python
df['postalcode'] = df['postalcode'].astype(str).str.replace('.0', '', regex=False) #removes trailing 0 in postal codes
```

```python
for col in ['status', 'country', 'dealsize', 'city']:
    if col in df.columns:
        df[col] = df[col].astype(str).str.title()  #standardize text fields
```

```python
df.to_csv("cleaned_sales_data.csv", index=False)

print("Cleaning complete!")
```
